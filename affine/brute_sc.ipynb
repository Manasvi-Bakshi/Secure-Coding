```python
def mod_inverse(a, m):
    for x in range(1, m):
        if (a * x) % m == 1:
            return x
    return None

def affine_decrypt(ciphertext, a, b):
    plaintext = ""
    a_inv = mod_inverse(a, 26)
    if a_inv is None:
        return None

    for char in ciphertext:
      if char.isalpha():
        y = ord(char.upper()) - ord('A')
        x = (a_inv * (y - b)) % 26  #Formula
        plaintext += chr(x + ord('A'))
      else:
        plaintext += char
    return plaintext


ciphertext = "FMXVEDKAPHFERBNDKRXRSREFMORUDSDKDVSHVUFEDKAPRKDLYEVLRHHRH"
# key a=3 and b = 5
possible_a = [a for a in range(1,26) if gcd(a,26)==1]

for a in possible_a:
    for b in range(26):
        decrypted_text = affine_decrypt(ciphertext, a, b)
        if decrypted_text:
            print(f"Key (a={a}, b={b}): {decrypted_text}")
```
